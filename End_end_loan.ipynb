```python
df.isnull().sum()
```

```python
### Dropping insignificant column/s

df=df.drop("Loan_ID",axis=1)  # axis=1 represent column
```

```python
df.shape
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
for col in df.select_dtypes(include=['number']).columns:
    median_value = df[col].median()
    df[col].fillna(median_value, inplace=True)
```

```python
df.isnull().sum()
```

```python
for col in df.select_dtypes(include=['object']).columns:
    mode_value = df[col].mode()[0]
    df[col].fillna(mode_value, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df["Dependents"].value_counts()
```

```python
df["Dependents"]=df["Dependents"].replace("3+",4)
```

```python
df["Dependents"].value_counts()
```

```python
df.dtypes
```

```python
df["Dependents"]=df["Dependents"].astype(int)
```

```python
### Dealing with categorical columns

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

for col in df.select_dtypes(include=['object']).columns:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
df.dtypes
```

```python
df.sample(25)
```

```python
from sklearn.model_selection import train_test_split   # to split the data package
from sklearn.linear_model import LogisticRegression  # Classification
#from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
X=df.drop(["Loan_Status"],axis=1)   # independent variables
y=df["Loan_Status"]                 # dependent variables
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model_rf=RandomForestClassifier()
```

```python
model_rf.fit(X_train, y_train)
```

```python
model_rf.score(X_train,y_train)
```

```python
model_rf.score(X_test,y_test)
```
